```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## Uniform Dist

```python

values = np.random.uniform(low=-10, high=10, size=10000)
plt.hist(values, 100)
plt.show()
```

```python
from scipy.stats import norm, expon, poisson
```

```python
x = np.arange(start=-3, stop=3, step=0.001)
x
```

```python
plt.plot(x, norm.pdf(x))
```

```python
# this time with mu and sigma
mu = 5
sigma = 2
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 150)
plt.show()
```

```python
x = np.arange(0, 10, 0.0001)
plt.plot(x, expon.pdf(x))
```

```python
mu = 500
x = np.arange(400, 600, 5.5)
plt.plot(x, poisson.pmf(x, mu))
```
