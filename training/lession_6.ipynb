```python
from __future__ import division
from datetime import datetime

# tables and math
import pandas as pd
from pandas import Series, DataFrame
import numpy as np

# visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# due to a bug
pd.core.common.is_list_like = pd.api.types.is_list_like
```

```python
from pandas_datareader import DataReader
from pandas_datareader import data as pdr
```

```python
tech_list = ["AAPL", "MSFT", "AMZN", "IBM"]
end = datetime(2018, 4, 4, 18, 00)
start = datetime(end.year - 1, end.month, end.day)
```

```python
symbols = {stock: DataReader(stock, "yahoo", start, end) for stock in tech_list}
```

```python
comp = "MSFT"
symbols[comp]
```

```python
symbols[comp].info()
symbols[comp].describe()
```

```python
df = symbols[comp]["Close"]
df=pd.to_numeric(df)
df.plot(legend=True, figsize=(10, 4))
```

```python
ma_day = [10, 20, 50]
for ma in ma_day:
    column_name = f"MA for {ma} days"
    symbols[comp][column_name] = symbols[comp]["Close"].rolling(window=ma).mean()
```

```python
symbols[comp][["Close", "MA for 10 days", "MA for 20 days", "MA for 50 days"]].plot(figsize=(10, 4))
```

```python
symbols[comp]["Daily Return"] = symbols[comp]["Close"].astype(float).pct_change()
symbols[comp]["Daily Return"].plot(figsize=(10,4), legend=True, linestyle="--", marker="o")
```

```python
sns.distplot(symbols[comp]["Daily Return"].dropna(), bins=100, color="orange")
```

```python
closing_df = DataReader(tech_list, "yahoo", start, end)["Close"]
```

```python
closing_df = pd.concat([symbols[c]["Close"] for c in tech_list], axis=1)
closing_df.columns = tech_list
```

```python
closing_df.describe()
```

```python
tech_rets = closing_df.pct_change()
```

```python
tech_rets.dropna().head()
```

```python
sns.jointplot(x="AMZN", y="MSFT", data=tech_rets, kind="scatter", height=10)
```

```python
sns.pairplot(tech_rets.dropna())
```

```python
returns_figs = sns.PairGrid(tech_rets.dropna())
returns_figs.map_upper(plt.scatter, color="purple")
returns_figs.map_lower(sns.kdeplot, cmap="cool_d")
returns_figs.map_diag(plt.hist, bins=50, color="blue")
```
