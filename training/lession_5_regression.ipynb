```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
print(boston.DESCR)
```

```python
plt.hist(boston.target, bins=50)
plt.xlabel("Price in $K", color="w")
plt.ylabel("# of Houses", color="w")
plt.show()
```

```python
for i in range(13):
    plt.scatter(boston.data[:,i], boston.target)
    plt.show()
```

```python
boston_df = DataFrame(boston.data)
boston_df.columns = boston.feature_names
boston_df.head(5)
```

```python
boston_df['Price'] = boston.target
```

```python
boston_df.head(1)
```

```python
sns.lmplot('RM', 'Price', data=boston_df)
```

```python
# X = boston_df.RM
Y = boston_df.Price
X = np.vstack(boston_df.RM)
X = np.array([[value, 1] for value in boston_df.RM])
```

```python
m, b = np.linalg.lstsq(X, Y)[0]
```

```python
plt.plot(boston_df.RM, boston_df.Price, "o")
x = boston_df.RM
plt.plot(x, x*m+b, "r", label="The best FIT")
```

```python
results = np.linalg.lstsq(X, Y)
error_total = results[1]
rmse = np.sqrt(error_total/len(X))
print(f"The RMSE of our model is: {rmse}")
```
